Join Data

```python
# Transaction & Point
TP = pd.merge(left = transaction_df,
              right = point_df[['ProductSKU','Points']],
              left_on ='ProductSKU',
              right_on = 'ProductSKU',
              how = 'left')
TP.drop(columns = ['ProductName'],inplace = True)
```

```python
TP.info()
```

```python
TP.head()
```

```python
#Transaction & User
TU = pd.merge(left=TP,
              right=user_df,
              left_on='UserId',
              right_on='UserId',
              how='left')
```

```python
TU.info()
```

```python
TU.head()
```

```python
# Transaction & Profile
FN = pd.merge(left=TU,
                 right=profile_df[['OwnerId', 'Distributor', 'Province']],
                 left_on='UserId',
                 right_on='OwnerId',
                 how='left')
FN.drop(columns = ['OwnerId'],inplace = True)
FN.drop(columns=['Distributor_y'],inplace=True)
FN.rename(columns={'Distributor_x':'Distributor'},inplace=True)
```

```python
FN.info()
```

```python
while FN['Province'].isnull().any():
    # Step 1: Identify rows with missing 'Province' and UserType = 'Employee'
    condition = (FN['UserType'] == 'Employee') & (FN['Province'].isnull())
    empty_province_indices = FN[condition].index

    # Step 2: Get 'GarageId' from FN
    garage_ids = FN.loc[empty_province_indices, 'GarageId']

    # Step 3: Filter user_df for UserType = 'Owner' and match 'GarageId'
    owner_df = user_df[(user_df['UserType'] == 'Owner') & (user_df['GarageId'].isin(garage_ids))]

    # Step 4: Extract 'UserId' from owner_df
    owner_ids = owner_df['UserId']

    # Step 5: Map 'UserId' to 'Province' in profile_df
    province_mapping = profile_df.set_index('OwnerId')['Province']
    provinces = owner_ids.map(province_mapping)

    # Step 6: Update 'Province' in FN
    FN.loc[empty_province_indices, 'Province'] = provinces
```
